```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\ilhan\Desktop\gray_img.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
x,y = img.shape
g = np.zeros((x,y), dtype=np.float32)
pepper = 0.1
salt = 0.95
```

```python
for i in range(x):
    for j in range(y):
        rdn = np.random.random()
        if rdn < pepper:
            g[i][j] = 0
        elif rdn > salt:
            g[i][j] = 1
        else:
            g[i][j] = img[i][j]
```

```python
g = cv2.cvtColor(g,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(g)
plt.title("Salt and Pepper Noise")
plt.axis("off")
plt.show()
```
