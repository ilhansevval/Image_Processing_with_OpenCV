# **Logarithmic Transformation**

Bir görüntünün logaritmik dönüşümü gray scale görüntü dönüşümlerinden biridir.

Görüntüde bulunan tüm piksel değerleri logaritmik değerleri ile değiştirilir.

Logaritmik dönüşüm görüntüdeki koyu pikselleri (düşük piksel değerli), açık piksellere kıyasla (yüksel piksel değerli) daha fazla yayılır ve bu pikseller açık pikseller olur. 

**Logaritmik Dönüşüm Grafiği**

image.png

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read image
img = cv2.imread(r"C:\Users\ilhan\Desktop\gray_img.png")
```

```python
c = 255 / np.log(1 + np.max(img))
log_transformation = c * (np.log(img + 1))
log_transformation = np.array(log_transformation, dtype=np.uint8)
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis("off")

plt.subplot(1, 2, 2) 
plt.imshow(log_transformation)
plt.title('Logarithmic transformation')
plt.axis("off")
plt.show()
```
