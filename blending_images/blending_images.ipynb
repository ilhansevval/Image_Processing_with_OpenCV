# **Blending Images**

Open CV ile iki görsel karıştırılarak tek bir görüntü elde edilebilir.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#read image
img1 = cv2.imread(r"C:\Users\ilhan\Desktop\img1.jpg")
img2 = cv2.imread(r"C:\Users\ilhan\Desktop\img2.jpg")
```

```python
#show image
plt.figure()
plt.imshow(img1)
plt.axis("off")

plt.figure()
plt.imshow(img2)
plt.axis("off")
plt.show()
```

imshow parametresi matplotlib kütüphanesinde görseller görselleştirilirken kullanılır.

Görselleştirilen fotoğraflar orijinal fotoğraflardan farklıdır.

Bunun sebebi Open Cv'ye görseller yüklenirken default olarak bgr (blue,green,red) renk skalasında yükler.

Orijinal görseller rgb formatındadır.

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
#show image
plt.figure()
plt.imshow(img1)
plt.axis("off")

plt.figure()
plt.imshow(img2)
plt.axis("off")
plt.show()
```

Görseller cvtColor (convert color) methoduyla bgr formatından rgb formatına çevrildiğinde orijinal görseller elde edilmiş olur.

Görseller birleştirilirken aynı boyutta olmalıdır. Çünkü farklı boyuttaki iki matris toplanamaz.

```python
print("İmage 1 Shape: ",img1.shape)
print("İmage 2 Shape: ",img2.shape)
```

Birinci görsel için 836 y eksenini, 1278 x eksenini 3 ise görselin renkli olduğunu gösterir.

Bakıldığında görsellerin genişlikleri ve yüksekliklerinin farklı olduğu görülür.

```python
#resize image
img1 = cv2.resize(img1,(600,600))
print("İmage 1 Shape: ",img1.shape)
img2 = cv2.resize(img2,(600,600))
print("İmage 2 Shape: ",img2.shape)
```

Görseller resize methodu ile aynı boyuta getirilmiştir.

```python
#show image
plt.figure()
plt.imshow(img1)
plt.axis("off")

plt.figure()
plt.imshow(img2)
plt.axis("off")
plt.show()
```

```python
#blended image = alpha*img1 + beta*img2
blended = cv2.addWeighted(src1 = img1, alpha = 0.5, src2 = img2, beta = 0.5, gamma = 0)
```

addWeighted methodu "blended image = alpha*img1 + beta*img2" bu formülün hazır bir şekilde yazılmış fonksiyonudur.

```python
plt.figure()
plt.imshow(blended)
plt.axis("off")
plt.show()
```

İki görselin tek bir görselde karıştırılmıştır.

alpha değeri 1, beta değeri 0 yapılsaydı sadece 1. görsel ile karşılaşılacaktı.
