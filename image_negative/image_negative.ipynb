# **Image Negative**

Görüntünde siyah renkler yoğunluktaysa, karanlık bölgelerde gömülü kalmış siyah ve beyaz detayları artırmak için kullanılabilir.

Negatif Dönüşüm fonksiyonu;

**s = L-1-r** şeklinde ifade edilir.

8 bitlik bir görüntü için L değeri 256'dır ve 0-255 arasındadır.

Görüntünün negatifini almak için L-1'den girişteki görüntünün yoğunluk değerleri çıkartılır.

Böylelikle görüntüdeki siyah renkler beyaz, beyaz renkler siyah olur.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread(r"C:\Users\ilhan\Desktop\mamogram.png",0)
```

```python
def negative(img):
    L=np.max(img) #255
    negative_img = L - img
    return negative_img
```

```python
negative_img = negative(img)
```

```python
stack_images = np.hstack((img,negative_img))
```

```python
plt.imshow(stack_images,cmap="gray")
plt.axis("off")
plt.title("Original vs. Negative Image")
plt.show()
```
