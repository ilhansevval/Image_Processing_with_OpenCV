# **Gradients**

Görüntü gradyanı, görüntüdeki yoğunluk (piksel değeri/intensity) veya renkteki yönlü bir değişikliktir.

Görüntü gradyanı kenar algılamada kullanılır.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#read image
img = cv2.imread(r"C:\Users\ilhan\Desktop\sudoku.jpg")
#görüntü 0 ile grayscale edilmiş şekilde okunur.
```

```python
#convert image bgr to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure()
plt.imshow(img)
plt.axis("off")
plt.title("Orijinal Görsel")
plt.show()
```

```python
#gradients on the x axis
sobelx = cv2.Sobel(img, ddepth = cv2.CV_16S, dx = 1, dy = 0, ksize = 5)
#ddepth parametresi output'un derinliğini gösterir.
#ksize -> filtrenin boyutu
#dx, x yönünde gradyan yapılacağından dolayı 1'dir.
#bu yüzden dy 0'dır.
```

```python
plt.figure()
plt.imshow(sobelx)
plt.axis("off")
plt.title("Sobel X")
plt.show()
```

Dik olan kenarlar tespit edilmiştir.

```python
#gradients on the y axis
sobely = cv2.Sobel(img, ddepth = cv2.CV_16S, dx = 0, dy = 1, ksize = 5)
```

```python
plt.figure()
plt.imshow(sobely)
plt.axis("off")
plt.title("Sobel Y")
plt.show()
```

Yatay olan kenarlar tespit edilmiştir.

Her iki yöndeki kenarlar birlikte tespit edilmek istendiğinde laplacian gradyan yöntemi kullanılır.

```python
#laplacian gradient
laplacian = cv2.Laplacian(img, ddepth = cv2.CV_16S)
```

```python
plt.figure()
plt.imshow(laplacian)
plt.axis("off")
plt.title("Laplacian")
plt.show()
```

Laplacian gradyan yöntemi ile her iki yöndeki kenarlar da tespit edilmiştir.
